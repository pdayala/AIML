### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
# Reading items csv file

df_movies_items = pd.read_csv("item.csv")

# Deleting one movie which has unknown rating and dropping the column 

df_movies_items.drop(df_movies_items.loc[df_movies_items['unknown'] == 1].index , inplace = True) 

df_movies_items.drop(columns='unknown',inplace=True)
```

```python
 # adding Release year 
    
df_movies_items['release year']=pd.DatetimeIndex(df_movies_items['release date']).year
 
```

```python
# dropping all other columns expect release year and genre's

df_movies_items.drop(columns=['movie id','movie title','release date'],inplace = True)

# setting release year as index

df_movies_items.set_index('release year',inplace=True)
```

```python
# Creating New Dataframe with aggregating the movies count for each year

df_movies_per_year = df_movies_items.groupby(['release year']).sum()
```

```python
# Reading items csv file to a new Dataframe used later to get total number of movies per release year

df_movies_item_2 = pd.read_csv("item.csv")

# Deleting one movie which has unknow rating and alos dropping the column 

df_movies_item_2.drop(df_movies_item_2.loc[df_movies_item_2['unknown'] == 1].index , inplace = True) 
```

```python
# Adding release year which is yyyy year part from release date 

df_movies_item_2['release year'] =   pd.DatetimeIndex(df_movies_item_2['release date']).year
```

```python
# Creating new DataFrame with only movie id , release year to aggregate and get sum 

df_movies_yr = df_movies_item_2[['movie id','release year']]
```

```python
df_movies_yr_cnt = df_movies_yr.groupby('release year')['movie id'].agg(['count'])
```

```python
# Joining 2 DataFrames 

df_movies_per_year_genre = df_movies_per_year.merge(df_movies_yr_cnt ,on = 'release year',how ='inner')
```

```python
# Creating new DataFrame with calculated percentages in place of count

df_movies_year_grp_percentage = (df_movies_per_year_genre.div(df_movies_per_year_genre['count'], axis= 0)* 100).round(2)
```

```python
# Dropping count which is not required in plotting 

df_movies_year_grp_percentage.drop(columns=['count'],inplace = True)
```

```python
# Ploting Head map for the results of percentages or Genre released per year..


plt.figure(figsize=(18,45)) 
sns.heatmap(df_movies_year_grp_percentage, annot = True,cmap= 'BuPu', fmt='g')
plt.show()
```

**Insights:**  Drama , Thrilar,Romance movies percentages has increased over the period , while wstern , war and animation has decreased.


```python
df_movies_year_grp_percentage.plot(figsize=(20,35),
                                   title="Genre Percentages of movie reease per year",
                                   xlabel='release year',
                                   ylabel='Genre Percentages');
```

```python
plt.figure(figsize=(18,45)) 
plt.hist(df_movies_year_grp_percentage)
plt.show()
```

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# 1. Find the count of ratings and average ratings for every movie.

df_count_mean_rating = df_rating_data.groupby('movie id').agg({'rating' : [np.size,np.mean]})
```

```python
#2. Slice the movies which have ratings more than 100.
df_filter_100 = df_count_mean_rating[df_count_mean_rating['rating']['size'] >100]
```

```python
#3. Sort values according to average rating such that movie which highest rating is on top & Select top 25 movies
df_top_25_rating =df_filter_100.sort_values([('rating','mean')], ascending=False)[:25]
```

```python
df_top_25_rating.columns = ['{}_{}'.format(x[0], x[1]) for x in df_filter_100.columns]
```

```python
#5. Merge function to get the movie titles
df_top_25_rating_movies_jn = df_top_25_rating.merge(df_movies_item,on='movie id',how = 'left')
```

```python
# Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times

df_top_25_rating_movies_jn[[ 'movie title','rating_size', 'rating_mean']]
```
